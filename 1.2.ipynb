```python
import numpy as np


weights = np.zeros(2)
bias = 0
learning_rate = 0.1


X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 1])


def step_activation(x):
    return 1 if x >= 0 else 0


print("Начальные веса: w1=0, w2=0, bias=0")
print("\n")
```

```python
for epoch in range(5):
    print(f"\nЭпоха {epoch + 1}:")
    errors = 0
    
    for i in range(len(X)):
        weighted_sum = np.dot(X[i], weights) + bias
        prediction = step_activation(weighted_sum)
        
        error = y[i] - prediction
        if error != 0:
            errors += 1
            weights += learning_rate * error * X[i]
            bias += learning_rate * error
            status = "обновление"
        else:
            status = "ок"
            
        print(f"  Пример {i+1}: {X[i][0]} OR {X[i][1]} = {prediction} "
              f"(ожидалось {y[i]}) - {status}")
        print(f"    Веса: w1={weights[0]:.2f}, w2={weights[1]:.2f}, bias={bias:.2f}")
    
    if errors == 0:
        print(f"\nОбучение завершено! Все примеры классифицированы верно.")
        break
```

```python
print("Финальные результаты:")
print(f"Веса: w1 = {weights[0]:.2f}, w2 = {weights[1]:.2f}")
print(f"Bias = {bias:.2f}")
print(f"Итоговая функция: y = step({weights[0]:.2f}*x1 + {weights[1]:.2f}*x2 + {bias:.2f})")
```

```python
print("\nПроверка:")
for x in X:
    result = step_activation(np.dot(x, weights) + bias)
    print(f"{x[0]} OR {x[1]} = {result}")
```
