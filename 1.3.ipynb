```python
import numpy as np
import random
```

```python
def create_user_item_matrix(n_users, n_items, interactions):
    matrix = np.zeros((n_users, n_items))
    
    for user_id, item_id, rating in interactions:
        if 0 <= user_id < n_users and 0 <= item_id < n_items:
            matrix[user_id, item_id] = rating
        else:
            print(f"Предупреждение: взаимодействие (user={user_id}, item={item_id}, rating={rating}) "
                  f"выходит за пределы матрицы {n_users}x{n_items}")
    
    return matrix

def generate_random_interactions(n_users, n_items, n_interactions, rating_range=(1, 5)):
    interactions = []
    for _ in range(n_interactions):
        user_id = random.randint(0, n_users - 1)
        item_id = random.randint(0, n_items - 1)
        rating = random.randint(rating_range[0], rating_range[1])
        interactions.append((user_id, item_id, rating))
    
    return interactions
```

```python
n_users = 5
n_items = 5
n_interactions = 12

random.seed(42)
interactions = generate_random_interactions(n_users, n_items, n_interactions)
```

```python
print("Сгенерированные взаимодействия:")
print("Формат: (user_id, item_id, rating)")
print('\n')
for i, (user, item, rating) in enumerate(interactions, 1):
    print(f"{i:2d}. Пользователь {user} оценил товар {item} на {rating}")

user_item_matrix = create_user_item_matrix(n_users, n_items, interactions)
```

```python
print("USER-ITEM матрица:")
print(f"Размер: {n_users} x {n_items}")
print("Строки - пользователи (0-4), Столбцы - товары (0-4)")
print('\n')

print("     ", end="")
for j in range(n_items):
    print(f"Товар {j:2d} ", end="")
print()

for i in range(n_users):
    print(f"Юзер {i:2d} | ", end="")
    for j in range(n_items):
        if user_item_matrix[i, j] > 0:
            print(f"  {user_item_matrix[i, j]:3.0f}  ", end="")
        else:
            print(f"  {user_item_matrix[i, j]:3.0f}  ", end="")
    print()
```

```python
print("Статистика матрицы:")
print(f"Всего взаимодействий: {n_interactions}")
print(f"Уникальных взаимодействий: {np.count_nonzero(user_item_matrix)}")
print(f"Плотность матрицы: {np.count_nonzero(user_item_matrix) / (n_users * n_items) * 100:.2f}%")
print(f"Минимальный рейтинг: {np.min(user_item_matrix[user_item_matrix > 0]) if np.any(user_item_matrix > 0) else 0}")
print(f"Максимальный рейтинг: {np.max(user_item_matrix)}")

try:
    import pandas as pd
    
    df = pd.DataFrame(
        user_item_matrix,
        index=[f"User_{i}" for i in range(n_users)],
        columns=[f"Item_{j}" for j in range(n_items)]
    )
    
    print("\n")
    print("USER-ITEM МАТРИЦА (Pandas DataFrame):")
    print(df)
except ImportError:
    print("\nPandas не установлен, пропускаем DataFrame представление")

def get_rating(matrix, user_id, item_id):
    if 0 <= user_id < matrix.shape[0] and 0 <= item_id < matrix.shape[1]:
        return matrix[user_id, item_id]
    else:
        return None
```

```python
print("Примеры запросов:")
test_pairs = [(0, 0), (2, 3), (4, 1), (1, 4)]
for user, item in test_pairs:
    rating = get_rating(user_item_matrix, user, item)
    if rating > 0:
        print(f"Рейтинг пользователя {user} для товара {item}: {rating:.0f}")
    else:
        print(f"Пользователь {user} не оценивал товар {item}")
```
